```python
import pysat
import datetime
```

```python
start = datetime.datetime(2009,1,1)
stop = datetime.datetime(2009,1,3)
date = datetime.datetime(2009,1,2)

info = {'index': 'mlt', 'kind': 'local time'}
ivm = pysat.Instrument(platform='cnofs', name='ivm', clean_level='dusty', orbit_info=info)
ivm.download(start, stop)
```

```python
ivm.load(date=date)
```

```python
ivm.orbits[9]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#depletions in a TID
plt.plot(ivm['mlt'], ivm['ionDensity'])
plt.xlabel('mlt')
plt.ylabel('ion density')
plt.title(date)
```

```python
#drifts associated with TID
plt.plot(ivm['mlt'], ivm['ionVelmeridional'])
plt.xlabel('mlt')
plt.ylabel('meridional drift')
plt.title(date)
```
